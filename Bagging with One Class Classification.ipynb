```python
from sklearn.model_selection import train_test_split
X = xy.iloc[:,xy.columns != "Y"].values
y = xy.iloc[:,xy.columns == "Y"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=True)
```

```python
colors = ["blue","green","red","cyan","magenta","yellow","black","black","white"]
bagging_grid = pd.DataFrame()
bagging_acc = pd.DataFrame()
start = time.time()

#c_counter = 0
for s in range(1,500,1):
    print("*"*10)
    print(s)
    print()

    clfs = []
    d = np.array_split(X_train[np.where(y_train==0)[0]], s)
    for i in range(s):
        #clf_name = "clf{}".format(i)
        #print(clf_name)
        clf = IsolationForest(n_jobs=-1)
        _ = clf.fit(d[i])
        clfs.append(clf)

    xx = np.linspace(-100, 100, 201, endpoint=True)
    yy = np.linspace(-100, 100, 201, endpoint=True)
    xv, yv = np.meshgrid(xx, yy)
    grid = pd.DataFrame({"X01":xv.reshape(-1),"X02":yv.reshape(-1)})

    for key,clf in enumerate(clfs):
        #print("Y"+str(key))
        grid["Y"+str(key)] = clf.predict(grid[["X01","X02"]])
        #g = sns.lmplot(x="X01", y="X02", hue="Y"+str(key), data=grid, fit_reg=False, legend=True)
    
    grid["Y"] = round((grid.iloc[:,(grid.columns!="X01") & (grid.columns!="X02")].sum(axis=1)+s)/(2*s))
    bagging_grid["Y{:04d}".format(s)] = grid["Y"]

    #g = sns.lmplot(x="X01", y="X02", hue="Y", data=grid, fit_reg=False, legend=True)
    #g = plt.contour(xv, yv, grid["Y"].values.reshape(xv.shape), colors=colors[c_counter])

    bagging_acc.loc[s,"# Bag"] = s
    bagging_acc.loc[s,"Test TN"] = np.count_nonzero(clf.predict(X_test[np.where(y_test==0)[0]])==1)
    bagging_acc.loc[s,"Test TP"] = np.count_nonzero(clf.predict(X_test[np.where(y_test==1)[0]])==-1)
    bagging_acc.loc[s,"Test FN"] = np.count_nonzero(clf.predict(X_test[np.where(y_test==1)[0]])==1)
    bagging_acc.loc[s,"Test FP"] = np.count_nonzero(clf.predict(X_test[np.where(y_test==0)[0]])==-1)
    bagging_acc.loc[s,"Test Specificity"] = bagging_acc.loc[s,"Test TN"]/(bagging_acc.loc[s,"Test TN"]+bagging_acc.loc[s,"Test FP"])
    bagging_acc.loc[s,"Test Recall"] = bagging_acc.loc[s,"Test TP"]/(bagging_acc.loc[s,"Test TP"]+bagging_acc.loc[s,"Test FN"])
    bagging_acc.loc[s,"Test Precision"] = bagging_acc.loc[s,"Test TP"]/(bagging_acc.loc[s,"Test TP"]+bagging_acc.loc[s,"Test FP"])
    bagging_acc.loc[s,"Test F1 score"] = 2/((1/bagging_acc.loc[s,"Test Recall"]) + (1/bagging_acc.loc[s,"Test Precision"]))

    
    #print("Train Class = 0 ACC: ",np.count_nonzero(clf.predict(X_train[np.where(y_train==0)[0]])==1)/y_train[y_train==0].shape[0])
    #print("Train Class = 1 ACC: ",np.count_nonzero(clf.predict(X_train[np.where(y_train==1)[0]])==-1)/y_train[y_train==1].shape[0])
    #print("Test Class = 0 ACC: ",np.count_nonzero(clf.predict(X_test[np.where(y_test==0)[0]])==1)/y_test[y_test==0].shape[0])
    #print("Test Class = 1 ACC: ",np.count_nonzero(clf.predict(X_test[np.where(y_test==1)[0]])==-1)/y_test[y_test==1].shape[0])
    #c_counter += 1
    
run = time.time() - start
print("Time use: {:2f}".format(run))
```
